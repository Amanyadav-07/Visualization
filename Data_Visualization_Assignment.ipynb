###5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

```python
np.random.seed(0)
num_samples = 1000
cov_matrix = [
    [1.0, 0.8, 0.0, -0.6],
    [0.8, 1.0, 0.3, 0.0],
    [0.0, 0.3, 1.0, 0.2],
    [-0.6, 0.0, 0.2, 1.0]
]
data = np.random.multivariate_normal(mean=[0, 0, 0, 0], cov=cov_matrix, size=num_samples)
df = pd.DataFrame(data, columns=['Feature 1', 'Feature 2', 'Feature 3', 'Feature 4'])
corr_matrix = df.corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', vmin=-1.0, vmax=1.0)
plt.title('Correlation Matrix of Features')
plt.show()
```

###Using the sales data, generate a heatmap to visulaize the variation in sales across different months and days.

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

heatmap_data = df.pivot_table(index='Month', columns='Day', values='Sales', aggfunc='mean')


fig = px.imshow(heatmap_data.values,
                labels=dict(x="Day", y="Month", color="Sales"),
                x=list(heatmap_data.columns),
                y=list(heatmap_data.index),
                color_continuous_scale='Viridis',
                title='Variation in Sales Across Months and Days')
fig.update_xaxes(side="top")
fig.show()
```

###4. Using the given x and y data, generate a 3D surface plot to visualize the function.

```python
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)
# Create 3D surface plot
fig = go.Figure(data=[go.Surface(z=z, x=x, y=y)])
fig.update_layout(title='3D Surface Plot of Function')
fig.update_layout(scene=dict(xaxis_title='X', yaxis_title='Y', zaxis_title='Z'))
fig.show()
```

#BOKEH ASSIGNMENT

###1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.


```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)
output_notebook()
p = figure(title="Sine Wave", x_axis_label='X', y_axis_label='Y')
p.line(x, y, line_width=2)
show(p)

```

###2. Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns

```python
np.random.seed(0)
n_points = 50
data = {
    'x': np.random.rand(n_points) * 10,
    'y': np.random.rand(n_points) * 10,
    'sizes': np.random.randint(5, 15, n_points),
    'colors': np.random.rand(n_points)
}
df = pd.DataFrame(data)
output_notebook()
p = figure(title="Scatter Plot", x_axis_label='X', y_axis_label='Y')
p.circle('x', 'y', size='sizes', color=bokeh.colors.RGB(0,0,0), alpha=0.6, source=df)
show(p)
```

###3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.


```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
output_notebook()
p = figure(title="Histogram of Random Data", background_fill_color="#fafafa")
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:],fill_color="navy", line_color="white", alpha=0.5)
p.xaxis.axis_label = 'Value'
p.yaxis.axis_label = 'Frequency'
show(p)

```
